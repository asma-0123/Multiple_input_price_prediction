```python
plt.title("area,bedrooms , age , vs price ")
plt.xlabel("area")
plt.ylabel("price")
plt.plot(df["area"], df["bedrooms"], label="bedrooms", color="blue", marker="o", markersize=8, mfc="yellow")
plt.plot(df["area"], df["age"], label="age", color="r", marker="s", markersize=8, mfc="pink")
plt.legend(loc="upper right")
plt.grid(color="orange")

plt.show()
```

```python
plt.title("area,bedrooms , age , vs price ")
plt.xlabel("area")
plt.ylabel("price")
plt.plot(df["area"], df["price"], label="price", color="blue", marker="o", markersize=8, mfc="yellow")

plt.legend(loc="upper right")
plt.grid(color="orange")
```

```python
plt.title("bedrooms ,age, vs price ")
plt.xlabel("bedrooms")
plt.ylabel("price")
plt.plot(df["bedrooms"], df["price"], label="price", color="blue", marker="o", markersize=8, mfc="yellow")
plt.plot(df["age"], df["price"], label="age", color="r", marker="s", markersize=8, mfc="pink")
plt.legend(loc="upper right")
plt.grid(color="orange")
```

```python
plt.scatter(df["area"],df["price"],color="red",ec="yellow",s=100,alpha=1)
plt.show()
```
